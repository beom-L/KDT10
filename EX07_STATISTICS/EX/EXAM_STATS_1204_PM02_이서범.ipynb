```python
# 5점
# ======================================================================
# Q18. (실습 — 전처리 기초)  
# ======================================================================
# `titanic.csv`를 불러와 다음을 수행하시오.

# 1) Age 결측치를 중앙값으로 대체하고, 결과를 `df_age_filled`라는 DataFrame에 저장  
# 2) Fare 항목의 히스토그램을 그리시오  
# 3) Fare에서 이상치(상위 1%) 값의 개수를 계산하여 `q18_outlier_count` 변수에 저장  

# 아래 코드 셀에 작성하세요.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

DATA_FILE = './titanic.csv'
df = pd.read_csv(DATA_FILE)

# 1 정답
# df['Age'].isna().sum() # 30개 있음
median_ = df['Age'].median(numeric_only=True)
df_age_filled = df.copy()
df_age_filled['Age'] = df_age_filled['Age'].fillna(median_)
display(df_age_filled)

# 2 정답
# plt.bar(df.index, df['Fare']) ## 처음에 bar라고 생각해서 x를 index로 넣었는데 hist사용하는게 맞음
plt.hist(df['Fare']) 
plt.show()

# 3 정답
ret = pd.Series(sorted(df['Fare']))
display(ret.describe())
# count    400.000000
# mean      43.747006
# std       27.213125
# min       15.053605
# 25%       22.375865
# 50%       35.094211
# 75%       52.140445
# max      123.221817
plt.boxplot(ret) ## 그림을 보니 이상치는 큰 값에만 존재함
q18_outlier_count = ret[ret > ret.quantile(0.99)].count()
q18_outlier_count
```
