```python
## 모듈 로딩
import pandas as pd
```

```python
## 데이터 준비
DATA_FILE = '../Data/titanic.csv'

## CSV => DataFrame 로딩
titanicDF = pd.read_csv(DATA_FILE, usecols=[0,2,3,6,8])                             ## 원하는 cols의 인덱스만 가져오는거임
titanicDF = pd.read_csv(DATA_FILE, usecols=['survived','sex','age','fare','class']) ## 이렇게 Label로도 들고올 수 있음  
```

[3] 그룹화 진행 <hr>

```python
# ## [3-1] class 컬럼으로 그룹화
# classDGB = titanicDF.groupby(['class'])  ## 그룹 객체 자체는 데이터를 직접 보여주지는 않지만, 각 그룹별로 통계 계산이나 데이터 접근을 할 수 있는 "그룹 객체" 상태가 됩니다.

# ## 그룹화 후 속성들
# print(f'그룹 수 : {classDGB.ngroups}개')        ## (예: 1등석, 2등석, 3등석 → 그룹 수 = 3)

# for k,v in classDGB.groups.items():
#     print(k,v)
#     print(classDGB.get_group(k).head())
# print(f'그룹별 데이터 행/열 수\n{classDGB.size()}')



##[3-1] class 컬럼으로 그룹화

# classDGB = titanicDF.groupby(['class'])
classDGB = titanicDF.groupby(['class'],as_index=False)

## 그룹화 후 속성들
print(f'그룹수 : {classDGB.ngroups}개')

for k in classDGB.groups:
    print(k)
    print(classDGB.get_group(k).head())

print(f'그룹별 데이터/행 수 \n{classDGB.size()}')
```

```python
def test(obj):
    print(type(obj))
    print(obj)
```

```python
## [3-2] 그룹별 집계
display(classDGB.mean(numeric_only=True))  ## 안에 object가 있어서, 숫자만 쓰겠다는걸 True로 해주면 계산 됨

display(classDGB.sum(numeric_only=True))

display(classDGB.agg(test))
```

```python
display(classDGB['fare'].sum(numeric_only=True))
display(classDGB['fare'].transform('sum'))

## 그룹별 연산 결과를 동일 DF/Series로 반환
titanicDF['total'] = classDGB['fare'].transform('sum')
titanicDF
```
