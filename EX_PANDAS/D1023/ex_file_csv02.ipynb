### [ 파일 입력과 출력 ]
- pandas.read_XXX() : 다양한 포맷의 파일 ==> DataFrame으로 변환
    * 파일 종류에 따라서 변환 시 설정 필요함 !
- pandas.to_XXX() : DataFrame => 다양한 포맷 str 또는 file 변환

```python
## [1] 모듈 로딩
import pandas as pd
import os           ## 해줘야지, 파일이나 폴더를 쓸 수 있음  ## 모듈, 경로, 폴더, 시스템 관련 처리 모듈
import glob
```

```python
## [2] 데이터 준비 - 파일 리스트 추출
DATA_DIR = '../Data'

FILE_LIST = os.listdir(DATA_DIR)
print(FILE_LIST)

## 파일 경로 + 파일명
FILE_NAME = f'{DATA_DIR}/{FILE_LIST[0]}'

print(FILE_NAME)
```

```python
## [3] 데이터 준비 - 파일 리스트 추출

## 폴더내 파일명을 리스트에 담아서 반환
## * 의미 : 모든 것
## *.* 의미 : 모든파일명.모든확장
FILE_LIST = glob.glob(f'{DATA_DIR}/*')
print(FILE_LIST)

FILE_LIST = glob.glob(f'{DATA_DIR}/*.*')
print(FILE_LIST)

FILE_LIST = glob.glob(f'{DATA_DIR}/*.csv') ## csv파일만 출력
print(FILE_LIST)

FILE_LIST = glob.glob(f'{DATA_DIR}/t*.csv') ## d로 시작하고 확장자가 csv파일만 출력
print(FILE_LIST)

FILE_NAME = FILE_LIST[1]
```

```python
## [3] CSV 파일 => DATAFrame 으로 변환

## 앞부분 일부 데이터 제거하고 가져오기: skiprows=51
csvDF=pd.read_csv(FILE_NAME,skiprows=51,header=None)

## 출력
csvDF
```

```python
## [4] CSV 파일 => DATAFrame 으로 변환

## 끝부분 일부 데이터 제거하고 가져오기: skipfooter=50
##                   추가로 앤진 설정 engine='python'
csvDF=pd.read_csv(FILE_NAME,skipfooter=50, engine='python')

## 출력
csvDF
```
