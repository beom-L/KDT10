19. 어떤 문제를 맞출 확률이 일정한 상황에서,  
학생이 8문제를 찍어서 풀었고 그중 3문제가 맞았다. 
이때 한 문제를 맞출 성공확률 p의 추정값을 구하시오.  


```python
from scipy.stats import binom
pro = 3/8
n = 8
ret = binom.pmf(3, n, pro) ## 성공한 횟수, 시행 횟수, 성공 확률
ret * 100
```

20. 한 카페에 손님이 평균적으로 1분당 2명 방문한다고 한다.
