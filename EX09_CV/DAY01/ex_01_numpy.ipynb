#### 【 Numpy - Array 생성 】

```python
## 모듈 로딩
import numpy as np
```

```python
## =================================================
## 빈 array 객체 생성
## =================================================
a1 = np.empty((3,4))
print(a1)
a2=np.empty(2)
a2
```

```python
## 초기화 => 값 넣기
a1.fill(255)
a1
```

```python
## =================================================
## 0으로 채운 array 객체 생성
## =================================================
## -> float64 기본
a1 = np.zeros((2,3))
print(a1.dtype, a1, sep='\n')

## -> int8 설정
a1 = np.zeros((2,3), dtype=np.int8)
print(a1.dtype, a1, sep='\n')

## -> bool 설정
a1 = np.zeros((2,3), dtype=np.bool)
print(a1.dtype, a1, sep='\n')
```

```python
## =================================================
## 1으로 채운 array 객체 생성
## =================================================
## -> float64 기본
a1 = np.ones((2,3))
print(a1.dtype, a1, sep='\n')

## -> int8 설정
a1 = np.ones((2,3), dtype=np.int8)
print(a1.dtype, a1, sep='\n')

## -> bool 설정
a1 = np.ones((2,3), dtype=np.bool)
print(a1.dtype, a1, sep='\n')
```
