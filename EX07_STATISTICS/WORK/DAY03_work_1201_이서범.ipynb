
#### 【 기술통계(상관관계) + 확률  】



---
[1] 두 변수 X, Y가 함께 증가하면 공분산은 어떤 값을 가지는가?  -> 1
1) 양수 2) 음수 3) 0 4) 항상 1

---

[2] 다음 중 상관계수의 특징으로 옳은 것은?  -> 3
1) 단위(scale)의 영향을 크게 받는다  
2) 값의 범위는 -∞ ~ +∞ 이다  
3) 단위가 달라도 비교가 가능하다  
4) 값이 0이면 반드시 관계가 없다  

---

[3] 피어슨 상관계수는 어떤 관계를 측정하는가?  -> 2
1) 비선형 관계  
2) 선형 관계  
3) 순위 기반 관계  
4) 범주형 데이터 기반 관계

---

[4] 스피어만 상관계수는 어떤 상황에서 피어슨보다 적합한가?  -> 1
1) 이상치가 많고 비선형 단조 관계일 때  
2) 두 변수 모두 범주형일 때  
3) 두 값이 모두 binary일 때  
4) 순위가 필요 없을 때  

---

[5] 켄달 τ(Tau) 순위 상관계수는 무엇을 비교하여 계산되는가?  -> 2
1) 평균과 분산  
2) 순위가 일치하는 쌍과 불일치 쌍  
3) 표준편차의 차이  
4) 범주형 빈도  

---

[6] 다음 중 Point-Biserial을 적용할 수 있는 경우는?  -> 3
1) 연속형 ↔ 연속형  
2) 순위형 ↔ 순위형  
3) 이진형 ↔ 연속형  
4) 범주형 ↔ 범주형  

---

[7] Cramér’s V는 어떤 데이터에 사용하는가?  -> 2
1) 연속형 ↔ 연속형  
2) 범주형 ↔ 범주형  
3) 연속형 ↔ 순위형  
4) 이진형 ↔ 연속형  

---

[8] 상관계수 r = -0.8 일 때 올바른 해석은?  -> 1
1) 매우 강한 음의 선형 관계  
2) 매우 약한 양의 관계  
3) 관계 없음  
4) 비선형 관계  

---

[9] 공분산이 0이면 무엇을 의미하는가?  -> 2
1) X와 Y는 독립이다  
2) 선형 관계가 없다  
3) 항상 상관계수도 0이다  
4) 두 변수의 관계를 알 수 없다  

---

[10] 상관관계가 높더라도, 반드시 인과관계가 있다고 말할 수 없다.  -> 참
1) 참 2) 거짓  

---

[11] 주사위를 한 번 던질 때 표본공간 Ω의 정의로 맞는 것은?  -> 2
1) {1}  
2) {1, 2, 3, 4, 5, 6}  
3) {홀수, 짝수}  
4) {성공, 실패}

---

[12] “주사위를 던져 짝수가 나오는 경우”는 어떤 개념인가?  -> 2
1) 표본공간  
2) 사건  
3) 확률변수  
4) 기댓값  

---

[13] 동전 2번 던져 “앞면의 개수”는 어떤 확률변수인가?  -> 1
1) 이산형  
2) 연속형  
3) 범주형  
4) 정규변수  

---

[14] 정규분포 특징으로 옳은 것은?  -> 3
1) 반드시 오른쪽으로 치우친 분포  
2) 평균과 중앙값이 다르다  
3) 평균을 중심으로 좌우대칭이다  
4) 이산형 분포이다  

---

[15] Z = (X – μ) / σ 의 목적은?  -> 4
1) 단위를 없애기 위해  
2) 새로운 평균과 분산을 만들기 위해  
3) 모든 정규분포를 동일한 기준으로 비교하기 위해  
4) 모두 맞음  

---

[16] 중심극한정리에 따르면, 표본평균의 분포는 어떻게 되는가?  -> 2
1) 항상 이항분포  
2) 표본수가 크면 정규분포에 가까워진다  
3) 반드시 균일분포  
4) 기댓값이 항상 0이다  

---

[17] 주사위에서 A={1,2,3} 이라면 Aᶜ는?  -> 2
1) {1,2,3}  
2) {4,5,6}  
3) {2,3,4}  
4) {1,6}
