#### 【 기대값과 분산】<hr>
<pre>
기대값 : 기대값 E[X] = 확률변수의 평균 : 이론적 평균
        실제 평균(X̄)은 시행 횟수가 늘수록 기대값에 가까워짐.

분  산 : Var(X)=E[(X−E[X])^2]
        확률변수 X가 얼마나 불확실하게 퍼져 있는지를 수치로 표현한 값
        즉, X의 흔들림(변동성)
        * 분산이 크다 → 값이 넓게 퍼져 있다 → 결과가 불확실
        * 분산이 작다 → 값이 모여 있다 → 결과가 안정적
        * 분산이 0   → 항상 같은 값 → 100% 확실
</pre>

```python
## ======================================================
## 모듈 로딩
## ======================================================
import random
import numpy as np
```

```python
## ======================================================
## 동전 던지기 횟수에 따른 변화 
## ======================================================
##=> 동전: 앞=1, 뒤=0
def toss():
    return random.choice([0,1])


##=> 시도 횟수n번에 따른 평균과 분산 변화
for n in [10, 100, 1000, 10000]:
    data = [toss() for _ in range(n)]
    print(f"{n:>5}번 시행 평균 = {sum(data)/n:.4f}", end=' ')
    print(f"{n:>5}번 분산 = {np.var(data):.4f}")
```

```python
## ======================================================
## 주사위 던지기 횟수에 따른 변화 
## ======================================================

##=> 주사위 1 ~ 6
def roll():
    return random.randint(1, 6)  


##=> 시도 횟수n번에 따른 평균과 분산 변화
for n in [10, 100, 1000, 10000]:
    data = [roll() for _ in range(n)]
    print(f"{n:>5}번 시행 평균 = {sum(data)/n:.4f}", end=' ')
    print(f"{n:>5}번 분산 = {np.var(data):.4f}")
```

```python
import numpy as np

# 1개 주사위 100만 번 시뮬레이션
N = 1_000_000
dice = np.random.randint(1, 7, size=N)  

# 기대값과 분산
print("E[X]             =", dice.mean())
print("Var[X]           =", dice.var())
print("Sample Var (n-1) =", dice.var(ddof=1))
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 300000

# 1, 2, 3개 주사위
dice1 = np.random.randint(1, 7, size=N)
dice2 = dice1 + np.random.randint(1, 7, size=N)
dice3 = dice2 + np.random.randint(1, 7, size=N)

# 결과 요약
print("1개  E,Var =", dice1.mean(), dice1.var())
print("2개  E,Var =", dice2.mean(), dice2.var())
print("3개  E,Var =", dice3.mean(), dice3.var())

# 시각화
plt.figure(figsize=(14,5))

plt.subplot(1,3,1)
plt.hist(dice1, bins=6, edgecolor='black')
plt.title("1 Dice")

plt.subplot(1,3,2)
plt.hist(dice2, bins=11, edgecolor='black')
plt.title("2 Dice")

plt.subplot(1,3,3)
plt.hist(dice3, bins=16, edgecolor='black')
plt.title("3 Dice")

plt.tight_layout()
plt.show()
```
