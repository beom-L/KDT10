### [ 웹 스크래핑 - 정보 추출 ]
- 수집된 웹 데이터에서 필요한 정보 추출하는 작업 : 스크래핑
- 스크래핑을 편리하게 진행하기 위한 패키지 : BeautifulSoup4
- binary/str 타입의 웹 데이터 ==> 객체화 (DOM)해서 프로그램 쉽게 사용할 수 있도록 함

[1] 모듈 로딩 <hr>

```python
from urllib.request import urlopen  ## web에서 URL 기반으로 다운로드
from bs4 import BeautifulSoup       ## web데이터 객체화 및 정보 추출용 모듈
```

[2] Markup 문자열에서 정보 추출<hr>
- find() 메서드
    * 매번 제일 첫번째 태그객체만 가져오는 문제 해결위한 메서드
    * 원하는 태그객체 추춣을 위한 다양한 설정이 가능


```python
## HTML파일에서 일부분만 선택 str로 저장
html_str = """
<body>
    <a class="ca main" href="http://www.naver.com">네이버</a>
    <a class="cb" href="http://www.google.com" target="_blank">구글</a> 
    <div style='border:2px solid black; background : #2df941;'>GOOD</div>
    <h1> BLUE VALENTINE</h1>
<P id='p1'>
You'll always be my &nbsp; blue valentine <BR>
You'll always be my blue valentine <BR>
</P>

<P id='p2'>
식어버린 너의 색은 blue <BR>
파랗게 멍이 든 my heart<BR>
몇 번이고 덧이 나 <BR>
열이 나 lovesick <BR>
이건 such a bad love <BR>
</P>

<P id='p3'>
We fight, we sigh, and stop <BR>
붉게 타오르다 한순간에 식어가 <BR>
Rewind, rewind, rewind <BR>
언제 그랬냐는 듯이 또 서롤 찾아 <BR>
</P>

<P id='p4'>
깊게 새긴 상처 <BR>
비친 red blood <BR>
부서진 forever <BR>
I can see it now <BR>
Can you see it now? <BR>
</P>
   

</body>
"""
```

```python
## markup str ==> 객체화
mbs = BeautifulSoup(html_str, 'html.parser')
```

```python
## 태그객체 전체 가져오기 => 제일 첫번쨰 태그만 추출
print("p 태그 전체 :", mbs.p)
print("p 태그 전체 :", mbs.p)
```

```python
## [1] find() 메서드 활용 : BS객체변수명.find(tag_str)
h1Tag = mbs.find('h1')

pTag = mbs.find('p')
```

```python
## [2] find() 메서드 활용 : BS객체변수명.find(tag_str, attr_dict)

## -> id가 p4인 p태그 추출
p4Tag = mbs.find('p', {'id':'p4'})
print(p4Tag)

## -> id가 p2인 p태그 추출
p2Tag = mbs.find('p', {'id':'p2'})
print(p2Tag.text, p2Tag.get_text()) ## 태그 빼고 글자만 가져오는거임
```

```python
# [2] find()메소드 활용 : BS객체변수명.find(tag_str, attr_dict)

# -> 두번째 a태그 객체 가져오기
a2Tag = mbs.find('a', {'class':'ca'})
print(a2Tag)
a2Tag = mbs.find('a', class_='ca')

# -> a태그의 속성값 추출 : 태그 속성 접근 : attrs
print('Tag의 모든 속성 읽기 :', a2Tag.attrs)
print('Tag의 모든 속성값 읽기 :', a2Tag.attrs.values())

# -> 두번째 a태그 객체 가져오기
a1Tag = mbs.find('a', {'class':'ca'})
a1Tag = mbs.find('a', class_='ca')

# -> a태그의 속성값 추출 : 태그 속성 접근 : attrs
print('Tag의 모든 속성 읽기 :', a1Tag.attrs)
print('Tag의 모든 속성값 읽기 :', a1Tag.attrs.values())
```

```python
## -> 두번째 a태그 객체 가져오기 : BS객체변수명.find(attr_dict)
a2Tag = mbs.find(attrs={'href':'http://www.naver.com'})
print('a2Tag :', a2Tag)

##          find(**kwargs)
a2Tag = mbs.find(href = 'http://www.naver.com')
print('a2Tag :', a2Tag)
```
