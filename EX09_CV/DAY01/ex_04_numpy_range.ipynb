#### 【 Numpy - 범위내 데이터 활용 ndarray 생성 】

- numpy.random.rand() : 0 ~ 1사이의 무작위 수로 생성 / randn() : 표준정규분포(평균:0, 분산:1)를 따르는 무작위 수로 생성 
- numpy.arange() : 순차적인 값으로 생성
- numpy.linspace() : 시작값~마지막값까지 num 등분 
- numpy.logspace() : 로그화 해줌

```python
## 모듈 로딩
import numpy as np      ## 이미지 데이터 저장 관련 모듈
import cv2              ## opencv 패키지
```

- 임의의 데이터 생성 및 형변환 
<hr>

```python
## numpy.random.rand
print(np.random.rand())
a = np.random.rand(2,3)
a
```

```python
## numpy.random.randn
print(np.random.randn())
b = np.random.randn(2,3)
b
```

```python
## arange
a = np.arange(5)
print(a, a.dtype, a.shape)

b = np.arange(5.0)
print(b, b.dtype, b.shape)

c = np.arange(3,9,2)
print(c, c.dtype, c.shape)
```

```python
## linspace
arr1 = np.linspace(2,3, num=3)
arr2 = np.linspace(1,5, num=5, endpoint=False)
arr3 , step=np.linspace(1,5, num=5, retstep=True)

arr3, step
```

```python
## logspace -> 상용로그화 해줌
arr1=np.logspace(2,3, num=2)
arr2=np.logspace(1,3, num=3, endpoint=False)
arr3=np.logspace(1,3, num=3, endpoint=True)

arr1, arr2, arr3
```
