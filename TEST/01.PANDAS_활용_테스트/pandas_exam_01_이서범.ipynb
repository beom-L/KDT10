<hr> 3. 조건 필터링과 정렬 (cars_mpg) <hr> 

- 3-1: mpg >= 30 인 행만 `q3_1_df`에.
- 3-2: manufacturer == 'audi' & year == 1999 인 행만 `q3_2_df`에.
- 3-3: horsepower 내림차순 정렬 결과를 `q3_3_sorted`에.
- 3-4: origin별 평균 mpg Series를 `q3_4_mean_mpg`에.
- 3-5: mpg 상위 5% 차량만 `q3_5_top5p`에.

```python
# === [3] 조건 필터링과 정렬 ===
q3 = pd.read_csv(r'C:\Users\qjadl\Desktop\KDT10\TEST\01.PANDAS_활용_테스트\EXAM_DATA\cars_mpg.csv')
q3_1_df = q3[q3.mpg >= 30]
display(q3_1_df)

q3_2_df = q3[(q3.manufacturer == 'audi') & (q3.year == 1999) ]
display(q3_2_df)

q3_3_sorted = q3.sort_values('horsepower')
display(q3_3_sorted)

q3_4_mean_mpg = q3[['origin','mpg']].groupby(by=['origin'],observed=True).mean()
display(q3_4_mean_mpg)

# q3_5_top5p = [q3.duplicated('model')]     ## 중복 모델이 없으니, sort정렬해서 상위 5개 인덱싱하기
q3_5_top5p = q3.sort_values('mpg',ascending=False)
n = len(q3_5_top5p.index)   # 총 개수 구하기
back = int(n * (5/100))     # 상위 5% 계산 진행 -> 1명 추출
q3_5_top5p = q3_5_top5p.iloc[:back]
q3_5_top5p
```

<hr> 4. 결측치 처리 (titanic_simple) <hr>

- 4-2: age 결측치를 평균으로 채운 DataFrame을 `q4_2_filled`에.
- 4-3: 결측치가 1개라도 있는 행 제거 DataFrame을 `q4_3_dropna`에.
- 4-4: 결측치 여부(True/False) DataFrame을 `q4_4_isna`에.
- 4-5: 처리 전후 age 평균을 튜플로 `q4_5_before_after`에.

```python
# === [4] 결측치 처리 ===
df = pd.read_csv(r'C:\Users\qjadl\Desktop\KDT10\TEST\01.PANDAS_활용_테스트\EXAM_DATA\titanic_simple.csv')
q4_1_na = []
for c in df.columns:
    q4_1_na.append( [df[c].isna().sum()] )
display(q4_1_na)

df1 = df.copy()
q4_2_filled = df1.fillna(df1.age.mean())
display(q4_2_filled)

df2 = df.copy()
q4_3_dropna = df2.dropna()
display(q4_3_dropna)

q4_4_isna = df.isna()
display(q4_4_isna)

## 결측치 처리를 평균을 채웠을때로 기준잡음 -> q4_2_filled 사용
q4_5_before_after = (df.age.mean(), q4_2_filled.age.mean())
q4_5_before_after
```

<hr> 5. 그룹화와 집계 (employees) <hr>

<hr> 7. 병합과 결합 (user_info, user_score) <hr>

- 7-1: 두 DF를 id로 내부 조인 -> `q7_1_inner`.
- 7-2: 동일 컬럼명 충돌시 접미사 적용 예시: (`suffixes=('_L','_R')`)로 머지 -> `q7_2_suffix`.
- 7-3: user_score를 id 기준으로 평균 점수를 구한 DF -> `q7_3_score_mean`.
- 7-4: user_info와 7-3 결과를 left join -> `q7_4_left`.
- 7-5: user_info와 user_score를 행 방향 concat -> `q7_5_concat` (ignore_index=True).

```python
# === [7] 병합과 결합 ===
df_info = pd.read_csv(r'C:\Users\qjadl\Desktop\KDT10\TEST\01.PANDAS_활용_테스트\EXAM_DATA\user_info.csv')
df_score = pd.read_csv(r'C:\Users\qjadl\Desktop\KDT10\TEST\01.PANDAS_활용_테스트\EXAM_DATA\user_score.csv')

q7_1_inner = pd.concat([df_info,df_score], join='inner',keys='id')
display(q7_1_inner)

q7_2_suffix = pd.merge(df_info, df_score,suffixes=('_L','_R'))
display(q7_2_suffix)

q7_3_score_mean = pd.merge(left=df_score, right=df_info, left_on=['id'], right_on=['id'])
q7_3_score_mean['mean score'] = q7_3_score_mean.score.mean()
display(q7_3_score_mean)

q7_4_left = pd.merge(left=df_info, right=q7_3_score_mean, left_on=['id'], right_on=['id'])
q7_4_left['mean score'] = q7_4_left.score.mean()
display(q7_4_left)


q7_5_concat = pd.concat([df_info,df_score],axis='index')
display(q7_5_concat)
```

<hr> 8. apply / lambda / 사용자 정의 함수 (apply_test) <hr>

- 8-1: score1의 제곱근 Series -> `q8_1_sqrt`.
- 8-2: 행 단위(axis=1)로 score1+score2 합 Series -> `q8_2_row_sum`.
- 8-3: name과 age를 결합한 full_name('Kim_25' 형식) Series -> `q8_3_fullname`.
- 8-4: age>=30 -> '성인' else '청년' Series -> `q8_4_age_group`.

```python
# === [8] apply / lambda / 사용자 정의 함수 ===
df = pd.read_csv(r'C:\Users\qjadl\Desktop\KDT10\TEST\01.PANDAS_활용_테스트\EXAM_DATA\apply_test.csv')

q8_1_sqrt = df.score1 ** (1/2)
display(q8_1_sqrt)

q8_2_row_sum = df[['score1','score2']].sum(axis=1)
display(q8_2_row_sum)

df2 = df.copy()
df2.age = df2.age.astype('str')
df2['full_name'] = df2.name + '_' + df2.age
q8_3_fullname = df2.full_name
display(q8_3_fullname)

df3 = df.copy()
df3['성인 혹은 청년'] = df3['age'].apply(lambda x: '성인' if x >= 30 else '청년')
q8_4_age_group = df3
display(q8_4_age_group)


def num_mean(df):
    ## 수치형 컬럼을 찾기
    df = df.select_dtypes(include=['number'])
    ## 수치형 컬럼의 평균 구하기
    return df.mean()

q8_5_num_mean = num_mean(df)
display(q8_5_num_mean)
```
