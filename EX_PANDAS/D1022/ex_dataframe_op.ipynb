### [ DataFrame 연산 ]
- Series와 동일한 원리로 연산
- 같은 위치에 있는 원소끼리 연산
- 종류
    * DataFrame 객체와 숫자 연산
    * DataFrame 객체와 DataFrame 객체 연산
    * DataFrame 객체와 연산메서드

```python
## 모듈
import pandas as pd
```

```python
# 데이터 프레임 생성
dataDF = pd.DataFrame([[10,20,30],
                    [40,50,60]])
dataDF
```

[2] DataFrame과 숫자 연산 <hr>

```python
## 원소 단위로 각각 계산 진행 ==> 결과

ret1 = dataDF + 17
ret2 = dataDF - 17
ret3 = dataDF * 17
ret4 = dataDF / 17
ret5 = dataDF % 17
ret6 = dataDF // 17

display(dataDF)
print(f'dataSR + 17 => \n{ret1}','\n')
print(f'dataSR - 17 => \n{ret2}','\n')
print(f'dataSR * 17 => \n{ret3}','\n')
print(f'dataDF / 17 => \n{ret4}','\n')
print(f'dataDF % 17 => \n{ret5}','\n')
print(f'dataDF // 17 => \n{ret6}','\n')
```

[3] DataFrame과 DataFrame 연산 <hr>
- 동일 인덱스 원소끼리 연산 수행

```python
## 새로운 Series 객체 생성
dataDF2 = dataDF.copy()

dataDF3 = dataDF.copy()
dataDF3.index = [0,2]

display(dataDF3)
```

```python
## => dataDF + dataDF 연산 : 동일 인덱스 원소끼리 연산


ret1 = dataDF + dataDF2
ret2 = dataDF  - dataDF2
ret3 = dataDF  * dataDF2
ret4 = dataDF  / dataDF2
ret5 = dataDF  % dataDF2
ret6 = dataDF  // dataDF2

display(dataDF)
print(f'dataDF + dataDF2 => \n{ret1}','\n')
print(f'dataDF - dataDF2 => \n{ret2}','\n')
print(f'dataDF * dataDF2 => \n{ret3}','\n')
print(f'dataDF / dataDF2 => \n{ret4}','\n')
print(f'dataDF % dataDF2 => \n{ret5}','\n')
print(f'dataDF // dataDF2 => \n{ret6}','\n')
```

```python
## => dataSR + dataSRR3 연산 : 동일 인덱스 원소끼리 연산

ret1 = dataDF + dataDF3
ret2 = dataDF  - dataDF3
ret3 = dataDF  * dataDF3
ret4 = dataDF  / dataDF3
ret5 = dataDF  % dataDF3
ret6 = dataDF  // dataDF3

display(dataDF)
print(f'dataDF + dataDF3 => \n{ret1}','\n')
print(f'dataDF - dataDF3 => \n{ret2}','\n')
print(f'dataDF * dataDF3 => \n{ret3}','\n')
print(f'dataDF / dataDF3 => \n{ret4}','\n')
print(f'dataDF % dataDF3 => \n{ret5}','\n')
print(f'dataDF // dataDF3 => \n{ret6}','\n')
```

[4] Series와 연산 메서드 <hr>

```python
## 인덱스가 같은 원소 없을 때 계산 불가 ==> 해결
ret1 =dataDF.add(dataDF3, fill_value=0)
ret2 =dataDF.sub(dataDF3, fill_value=0)
ret3 =dataDF.mul(dataDF3, fill_value=0)
ret4 =dataDF.div(dataDF3, fill_value=0)
ret5 =dataDF.mod(dataDF3, fill_value=0)


display(dataDF)
print(f'dataDF + dataDF3 => \n{ret1}','\n')
print(f'dataDF - dataDF3 => \n{ret2}','\n')
print(f'dataDF * dataDF3 => \n{ret3}','\n')
print(f'dataDF / dataDF3 => \n{ret4}','\n')
print(f'dataDF % dataDF3 => \n{ret5}','\n')
# print(f'dataDF // dataDF3 => \n{ret6}','\n')
```
