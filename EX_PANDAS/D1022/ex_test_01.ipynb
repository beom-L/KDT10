### 【 행/열 선택 실습 】

[1] 모듈 로딩 및 데이터 준비 <hr>

```python
# 모듈 로딩
import pandas as pd
```

```python
# 데이터 준비
DATA_FILE1 = '../Data/data01.csv'
DATA_FILE2 = '../Data/data02.csv'
```

```python
# 데이터 로딩 및 DataFrame 변환 저장
dataDF1 = pd.read_csv(DATA_FILE1, header=None)

#-> index_col 매개변수 : 파일에서 특정 컬럼을 행인덱스로 설정
dataDF2 = pd.read_csv(DATA_FILE2, index_col=0)
```

```python
display(dataDF1, dataDF2)
```

[dataDF1 실습 - 행과 열 선택] <hr>

```python
## -------------------------------------------------------------
## 행 선택 => DF.iloc[위치인덱스]/DF.loc[라벨인덱스]
## -------------------------------------------------------------
## (1) 마지막행 선택
one_row1=dataDF1.iloc[-1]
one_row2=dataDF1.loc[7]

print("=================")
print("==> one_row")
print("=================")
display(one_row1)
display(one_row2)

## (2) 2번,3번,4번 행 선택 => List 방식
three_row1 = dataDF1.iloc[[2,3,4]]
three_row2 = dataDF1.loc[[2,3,4]]

print("=================")
print("==> three_row")
print("=================")
display(three_row1)
display(three_row2)

## (3) 홀수번째 행 선택 => slicing 방식
three_row1 = dataDF1.iloc[1:8:2]
three_row2 = dataDF1.loc[1:7:2]

print("=================")
print("==> odd_rows")
print("=================")
display(three_row1)
display(three_row2)
```

```python
## -------------------------------------------------------------
## 열 선택 : DF["컬럼명"], DF.컬럼명 <=str일때만
## -------------------------------------------------------------
## (1) 첫번째 컬럼 선택
one_col1=dataDF1[0]
#one_col2=dataDF1.0

print("=================")
print("==> one_col")
print("=================")
display(one_col1)

## (2) 2번,3번째 컬럼 선택 => List 방식
two_cols=dataDF1[[1,2]]

print("=================")
print("==> two_cols")
print("=================")
display(two_cols)

## (3) 2번,3번째 컬럼 선택 => slicing 방식
two_cols1 = dataDF1.iloc[:, 1:3]
two_cols2 = dataDF1.loc[:, 1:2]

print("========================")
print("==> two_cols BY Slicing")
print("========================")
display(two_cols1)
display(two_cols2)
```

[dataDF2 실습 - 행과 열 선택] <hr>

```python
## -------------------------------------------------------------
## 행 선택 => DF.iloc[위치인덱스]/DF.loc[라벨인덱스]
## -------------------------------------------------------------
## (1) 마지막행 선택
one_row1=dataDF2.iloc[-1]
one_row2=dataDF2.loc[10]

print("=================")
print("==> one_row")
print("=================")
display(one_row1)
display(one_row2)

## (2) 2번,3번,4번 행 선택 => List 방식
three_row1 = dataDF2.iloc[[2,3,4]]
three_row2 = dataDF2.loc[[5,6,7]]

print("=================")
print("==> three_row")
print("=================")
display(three_row1)
display(three_row2)

## (3) 홀수번째 행 선택 => slicing 방식
three_row1 = dataDF2.iloc[1:8:2]
three_row2 = dataDF2.loc[4:10:2]

print("=================")
print("==> odd_rows")
print("=================")
display(three_row1)
display(three_row2)
```

```python
## -------------------------------------------------------------
## 열 선택 : DF["컬럼명"], DF.컬럼명 <=str일때만
## -------------------------------------------------------------
## (1) 첫번째 컬럼 선택
one_col1=dataDF2["v1"]
#one_col2=dataDF2.v1

print("=================")
print("==> one_col")
print("=================")
display(one_col1)

## (2) 2번,3번째 컬럼 선택 => List 방식
two_cols=dataDF2[['v2','v3']]

print("=================")
print("==> two_cols")
print("=================")
display(two_cols)

## (3) 2번,3번째 컬럼 선택 => slicing 방식
two_cols1 = dataDF2.iloc[:, 1:3]
two_cols2 = dataDF2.loc[:, 'v2':]

print("========================")
print("==> two_cols BY Slicing")
print("========================")
display(two_cols1)
display(two_cols2)
```
