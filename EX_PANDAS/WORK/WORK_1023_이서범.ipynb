```python
## 모듈 생성
import pandas as pd
import numpy as np
```

```python
## 문제 1
data = [["2차전지(생산)", "SK이노베이션", 10.19, 1.29],
        ["해운", "팬오션", 21.23, 0.95],
        ["시스템반도체", "티엘아이", 35.97, 1.12],
        ["해운", "HMM", 21.52, 3.20],
        ["시스템반도체", "아이에이", 37.32, 3.55],
        ["2차전지(생산)", "LG화학", 83.06, 3.75]
]
columns = ["테마", "종목명", "PER", "PBR"]

#  (1-1) 종목명, PER, PBR 컬럼만 선택 후 DataFrame으로 저장하는 코드 작성하세요.
dataDF = pd.DataFrame(data,columns=columns)
dataDF1 = dataDF.iloc[:, 1:]
display(dataDF1)

#  (1-2)  PER이 20미만, 50이상인 데이터만 저장하는 코드 작성하세요.
mask = (dataDF['PER'] < 20) | (dataDF['PER'] >= 50)
dataDF2 = dataDF[ mask ]
display(dataDF2)

#  (1-3) 종목명을 행인덱스로 설정하는 코드 작성하세요.
mask_row = dataDF['종목명']
dataDF3 = dataDF.set_index(mask_row)
display(dataDF3)

#  (1-4) PBR 컬럼의 데이터를 기준으로 내림차순 정렬해 주세요.
dataDF4 = dataDF.sort_values('PBR')
display(dataDF4)

#  (1-5) PER, PBR 순서로 오름차순 정렬해 주세요.
dataDF5 = dataDF.sort_values(by=['PER','PBR'])
display(dataDF5)
```

```python
## 문제 2
data = [
    ["037730", "3R", 1510, 7.36],
    ["036360", "3SOFT", 1790, 1.65],
    ["005760", "ACTS", 1185, 1.28],
 ]
columns = ["종목코드", "종목명", "현재가", "등락률"]

dataDF = pd.DataFrame(data, columns=columns)
display('2.',dataDF)

# (2-1) "종목코드"를 행 인덱스로 설정하세요.
dataDF1 = dataDF.set_index('종목코드')
display('2-1',dataDF1)

# (2-2) 아래 결과가 출력되도록 코드 작성하세요.
# 종목명현재가
# 종목코드
# 037730 3R 1510
#  036360 3SOFT 1790
display('2-2',dataDF1.iloc[:-1])

#  (2-3) 현재가가 1,500원 이상이면서 1,700원 미만인 값만 출력하세요.
mask_price = (dataDF['현재가'] >= 1500) & (dataDF['현재가'] < 1700)
dataDF3 = dataDF[ mask_price ]
display('2-3',dataDF3)

#  (2-4) 목표가는 현재가에서 10% 상승한 가격으로 목표가 컬럼을 추가하세요.
dataDF['목표가'] = dataDF["현재가"]*(110/100)
display('2-4',dataDF)

#  (2-5) 아래 LG 데이터를 추가하세요.
# # 종목코드 종목명   현재가 등락률
# # 066570  LG전자  60000  3.84
dataDF1.loc['066570'] = ['LG전자', 60000, 3.84]
display('2-5',dataDF1)

```

```python
## 5 - 1 번
FILE_NAME = '../Data/titanic.csv'
data = pd.read_csv(FILE_NAME)
display('5.',data)

#  ▶ 성별이 여자인 승객 행만 필터링.
mask_1 = (data['sex'] == 'female')
display('5-1 ▶ 성별이 여자인 승객 행만 필터링',data[mask_1]) 

#  ▶ pclass가 1등석인 승객만.
mask_2 = (data['pclass'] == 1)
display('5-2 ▶ pclass가 1등석인 승객만',data[mask_2]) 

#  ▶ age가 10세 미만인 어린이만.
mask_3 = (data['age'] < 10)
display('5-3 ▶ age가 10세 미만인 어린이만',data[mask_3]) 

#  ▶ fare가 100이상인 고가 요금 승객만.
mask_4 = (data['fare'] >= 100)
display('5-4 ▶ fare가 100이상인 고가 요금 승객만',data[mask_4]) 
```
