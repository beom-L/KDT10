```python
import pandas as pd
import numpy as np
import os
```

```python
## 데이터 생성
data = {'나이' : [10, 20 ,30], '성별':['남','여','남'], '이름':['영민', '기범', '돌수']}
index_ = ['1번', '2번', '3번']
df = pd.DataFrame(data, index=index_)
display(df)
```

```python
## 행/열 삭제 ##
# 행 삭제
df2 = df.drop(['두번째','첫번째'])
display(df2)

# 열 살제
df3 = df.drop('성별', axis=1)
display(df3)
df3 = df.drop(columns='성별')
display(df3)
```

```python
## 산술 연산 (시리즈) ##
dfNumber = pd.Series({'국어':100, '영어':80, '수학':90})
dfNumber2 = pd.Series({'영어':80, '국어':80, '수학':100})

#1 시리즈 vs 숫자
percentage = dfNumber/100
display(dfNumber, percentage)

#2 시리즈 vs 시리즈  (+, -, *, /)
a = dfNumber - dfNumber2  ## 같은 과목명을 찾아서 정렬한 후 계산
display(a)

b = dfNumber - dfNumber2
c = dfNumber * dfNumber2
d = dfNumber / dfNumber2

result = pd.DataFrame([a,b,c,d], index=['뺄셈', '덧셈', '곱셈', '나눗셈'])
display(result)

#3 연산 메소드
## 연산 메소드 사용 후, NaN 결과는 fill_value 로 미리 대비하기
sr_add = dfNumber.add(dfNumber2, fill_value=0)
sr_sub = dfNumber.sub(dfNumber2, fill_value=0)
sr_mul = dfNumber.mul(dfNumber2, fill_value=0)
sr_div = dfNumber.div(dfNumber2, fill_value=0)
result = pd.DataFrame([sr_sub,sr_add,sr_mul,sr_div], index=['뺄셈', '덧셈', '곱셈', '나눗셈'])
display(result)
```

```python
## 산술 연산 (데이터프레임) ##
DATA_FILE = '../Data/titanic.csv'
data = pd.read_csv(DATA_FILE)
df = data.loc[:, ['age','fare']]
display(df.head()) ## 첫 5행만 출력

## 데이터프레임 vs 숫자
df2 = df + 10
display(df2.head())

## 데이터프레임 vs 데이터프레임
df3 = df2 - df
display(df3.head())

## 연산 메서드 사용
df3 = df.add(df2, fill_value=0)
df3
```

```python
## 필터링 ##
## 데이터 생성
sample1 = df.head(10)

# 조건식 적용
sample1['age'] < 20              ## 만족하는 값만 True 반환

# 불린 인덱싱
# 일반적인
sample1[ sample1['age'] < 20 ]   ## 만족하는 값만 반환
# loc 사용
sample1.loc[ sample1['age'] < 20 ]

# Not 연산자
sample1[ ~(sample1['age'] < 40)  ]

# 특정 행에서 조건식 걸기
maske = (data.age < 20) | (data.age > 10)
sample1.loc[ maske ]

maske = (data.age < 20) | (data.age > 10)
data.loc[ maske, ['age', 'sex', 'alone'] ]  ## 해당 범위에 속하는 age를 가진 인원의, age / sex / alone 열만 선택

# 새로운 열 추가
data['age_under_20'] = data.age < 20  ## 나이가 20 이하인 사람을 새로운 열에 추가
data
```

```python
## query() 메소드
# : 그냥 ''안에 원하는 조건식 입력하면됨
sample1.query('age < 20')

sample1.query('age<20 and fare > 25')
```

```python
## isin() 메소드
# : 안에 있는지 확인하고 T/F 반환
isinF = data['embark_town'].isin(['Southampton', 'Queenstown']) ## embark_town 안에 Southampton, Queenstown 키워드 있으면 True 반환
isinF
```
