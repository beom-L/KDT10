###【 그룹화 연산 】

[1] 모듈 로딩 및 데이터 준비 <hr>

```python
## 모듈 로딩
import pandas as pd
```

```python
## DataFrame 인스턴스 생성
DATA_FILE = '../Data/상품_매출.csv'

df = pd.read_csv(DATA_FILE)

## 구조 확인
df.info()
display( df )
```

[2] 그룹화 진행 및 기본 속성들 <hr>

```python
## 지점별 그룹화 
g = df.groupby('지점')

print(type(g))
```

```python
## 그룹 속성 확인
print(f"-> 그룹수 : { g.ngroups}개")        # 그룹 수
print(f"-> 그룹별" )                        # 각 그룹의 인덱스
for k, v in g.groups.items():
    print(k, v)
print(f"-> 그룹별 행 수\n{g.size()}" )       # 그룹별 행 수
```

```python
## 그룹별 데이터 추출
for k in g.groups.keys():
    print(f'\n[{k}]=====')
    print(g.get_group(k))       
```

[3] 그룹별 집계 메서드 <hr>

```python
## (1) 그룹별 합계
print("합계===>\n", g.sum(numeric_only=True))

## (2) 평균
print("\n평균===>\n", g.mean(numeric_only=True))
print()

## (3) 여러 집계 동시 적용
print(g.agg({'매출':['sum','mean'], '수량':['max','min']}))
```

```python
## 지점의 상품별 매출 합계와 평균 
for key in g.groups.keys():
    print(f'\n[{key}]=====')
    localDF = g.get_group(key).drop(columns='지점').reset_index(drop=True)
    
    localDF=localDF.groupby('상품')
    display( localDF.agg({'매출':['sum','mean']}) )
```

```python
## 지점별 상품에 따른 그룹화
g = df.groupby(['지점','상품'])

## 지점별 상품에 따른 매출 합계와 평균 집계 
display(g.agg({'매출':['sum','mean'], '수량':['mean','max']}))

## 그룹별 데이터 통계 요약
display(df.groupby('지점')['매출'].describe()) 

```

```python
## 각 지점 내에서 평균 대비 매출비율
df['매출비율'] = df.groupby('지점')['매출'].transform(lambda x: x / x.mean())
display(df.head())
```

```python
# 평균 매출이 200 이상인 지점만 남기기
partDF   = df.groupby('지점')
display( partDF['매출'].mean() )

filterDF = partDF.filter(lambda x: x['매출'].mean() >= 200)
display(filterDF.head())
```

```python
## ==========================================================
## 그룹별 가장 매출이 높은 행 (idxmax)
## ==========================================================
idx = df.groupby('지점')['매출'].idxmax()
print(idx, end='\n\n')

print(df.loc[idx, ['지점','상품','월','매출']])
```

```python
## ==========================================================
## 그룹별 상관관계
## ==========================================================
# 각 지점 내에서 매출과 수량의 상관계수
corr = df.groupby('지점')[['매출','수량']].corr().iloc[0::2, -1]
print(corr)
```
