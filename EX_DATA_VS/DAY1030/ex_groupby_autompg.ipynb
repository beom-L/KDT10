[3] 컬럼들의 관계 체크 <hr>
- 주제 : 연비에 영향이 있는 요소/요인을 분석
- 타겟 : 연비 컬럼 <= 다른 컬럼들의 영향으로 값이 결정. 종속변수컬럼
- 피쳐 : 타겟 컬럼에 영향을 끼치는 컬럼들. 독립변수컬럼   [ Feature , 독립변수 ]
- 컬럼들의 관계 정도 체크 => corr()

```python
## mpg컬럼에 대한 다른 컬럼들과의 관계 수치값
mpgcorr = mpgDF.corr(numeric_only=True)['mpg']

# mpgcorr.sort_values().abs() >= 0.7  ## T/F로 반환   # ! -는 상관관계가 어떻게 되는지 -> 절댓값 1에 가까울수록 관계가 있는거임 -> -1이면 음의 상관관계 +1이면 양의 상관관계
mpgcorr = mpgcorr.sort_values().abs()
mpgcorr = mpgcorr[mpgcorr>=0.7]  ## 총 4개정도 연관이 있음을 알 수 있음
print(mpgcorr)
print(mpgcorr.index)
```

```python
## mpg 컬럼과 관련성이 높은 컬럼들의 시각화
## plot()
# mpgDF.plot(x='weight', y='mpg', kind='scatter') ## weight와 mpg의 상관관계를 그래프로 그림

for col in mpgcorr.index[:-1]: ## -1 까지 주는 이유는 마지막 columns은 mpg니깐 자기 자신이라서 제외한거임
    mpgDF.plot(x=col, y='mpg', kind='scatter', title=f'corr - {mpgcorr[col]}') ## title은 
```
