### 【 윤동주 시인의 시 제목과 내용 추출 및 저장 】
- csv 파일로 저장


```python
## ============================================================
## 모듈 로딩
## ============================================================
from urllib.request import urlopen     # URL 해당 페이지 -> 로딩 HttpResponse Object
from urllib.request import Request     # 요청 객체 즉, RequestMessage 생성
from bs4 import BeautifulSoup          # 받은 HTML 데이터 -> 객체화 변환 모듈    
import pandas as pd 
```

```python
WIKI_URL = 'https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC?page=2&sort=title#section1'

req = Request(WIKI_URL, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'})
```

```python
html = urlopen(req)
htmlBS = BeautifulSoup(html, 'html.parser')
htmlBS
```

```python
titles=[]
Href = []

s1 = htmlBS.select('div.mw-content-ltr.mw-parser-output ul > li')

for i in s1:
    Href.append(i.a['href'])
    titles.append(i.get_text(strip=True))

print(titles)
```

```python
ass = 'https://ko.wikisource.org'

result_content = []
titles =  titles[3:]
Href = Href[3:]
for h in range(len(Href)):
    
    url = ass + Href[h]
    
    req = Request(url, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'})
    html = urlopen(req)
    htmlBS = BeautifulSoup(html, 'html.parser')
    a = htmlBS.select('div.prp-pages-output p')[1]
    ret = a.get_text(strip=True) if a else None
    result_content.append({"titles":titles[h], "ret": ret})
    print(ret)
    print("\n")
```

```python
pd.DataFrame(result_content).to_csv('final.csv')
```
