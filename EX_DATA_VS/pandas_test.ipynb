```python
## 모듈
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
## 파일 읽어오기 -> O
vgDF = pd.read_csv('./DATA/vgsales.csv',index_col='Rank')
vgDF.head()
```

```python
## 원본 프레임을 'Global_Sales' 컬럼을 기준으로 내림차순 정렬하세요.
vgDF2 = vgDF.copy()

# ## 이건 틀림 -> 왜냐하면 series로 만들고 정렬하는거니깐 데이터프레임이 아님
# vgDF2['Global_Sales'].sort_values()

## 이게 정답 !
vgDF2 = vgDF2.sort_values(by='Global_Sales', ascending=False)
vgDF2.head()
```

```python
## 원본 데이터프레임을 'Platform'과 'Global_Sales'를 기준으로 정렬하세요.  -> O
# 'Platform'은 오름차순으로 정렬하고 'Global_Sales'는 내림차순으로 정렬하세요.
vgDF3 = vgDF.copy()

vgDF3.sort_values(by=['Platform','Global_Sales'], ascending=[True, False])
```

```python
## 원본 데이터 프레임에서 Year 컬럼의 값이 NaN인 로우를 제거하고 이를 원본에 반영하세요. -> X
vgDF4 = vgDF.copy()

# ## 계속 시리즈로 뽑아내서 하는데, 데이터 프레임에 해야됨. 정신차려
# vgDF4.Year.dropna(inplace=True)

## 정답
vgDF4.dropna(subset='Year', inplace=True)
display(vgDF4)
```

```python
## 데이터프레임에서 'Year' 컬럼의 타입을 int로 변경하세요. -> O
vgDF4.Year = vgDF4.Year.astype('int')
display(vgDF4)
vgDF4.info()
```

```python
## 데이터프레임에서 Genre 컬럼의 각 장르별 개수를 출력하세요. -> X
# print(vgDF4.Genre.nunique())  -> 이거 아님

## 정답
vgDF4.Genre.value_counts()
```

```python
## 데이터프레임에서 value_counts 메서드를 사용해서 Genre 컬럼의 각 장르별 백분율을 계산하세요.
vgDF4.Genre.value_counts(normalize=True) * 100
```

```python
## 데이터프레임에서 게임 이름(Name)에 'Super Mario'가 포함된 게임들만 선택하세요.
mario = vgDF4.Name.str.contains('Super Mario') 
vgDF4 = vgDF4[mario]
vgDF4.head()
```

```python
## 데이터프레임의 groupby 메서드를 사용하여 플랫폼과 장르별 Global_Sales의 합을 계산하세요.
vgDF4[['Platform','Genre','Global_Sales']].groupby(by=['Platform','Genre']).sum(numeric_only=True)
```

```python
## 판다스 melt 함수를 사용하여 다음과 같은 데이터프레임을 생성하세요.
pd.melt(
    vgDF, 
    id_vars=['Name', 'Platform', 'Genre'],                              ## id_vars      : 그대로 들고 가는 것
    value_vars=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'],     ## value_vars   : 세로로 길게 펼칠 컬럼들
    var_name='Region',                                                  ## var_name     : 열을 세로로하면 해당 열을 어떻게 부를지 
    value_name='Sales'                                                  ## value_name   : 해당 셀에 들어있었던 수치값 이름
)
```

### 부동산 <hr>

```python
df_m = pd.read_excel('./DATA/매매_41133_202101.xlsx', index_col=0)
df_m.head()
```

```python
## 거래금액 컬럼에는 문자열(object) 타입으로 데이터가 저장돼 있습니다. -> O
# 콤마를 제거하고 데이터 타입을 숫자로 변경하세요.
df_m.거래금액 = df_m.거래금액.str.replace(',','').astype('int')
df_m.info()
df_m.head()
```

```python
## 다양한 크기의 아파트 가격을 비교하기위해 매매평단가를 계산합니다. 
#  일반적으로 평단가는 거래금액과 공급면적으로 계산하지만, 
#  이번문제에서는 주어진 데이터 거래금액과 전용면적으로 매매평단가를 계산해 봅시다.
#  m^2 를 평으로 환산할 때는 3.3으로 값을 나눕니다.
df_m['매매평단가'] = df_m['거래금액'] / (df_m['전용면적'] / 3.3) 
df_m.head()
```

```python
## 주어진 성남시 데이터에는 총 8개의 법정동이 존재합니다. 동별로 매매평단가의 평균을 계산하세요.
df1 = df_m[['법정동','매매평단가']].groupby(by=['법정동'], observed=True).mean(numeric_only=True)
df1
```

```python
# 하대원동의 아파트별 매매평단가를 출력하세요.
dfk = df_m[df_m['법정동'] == '하대원동']
dfk = dfk.sort_values('아파트')
dfk = dfk.loc[:, ['아파트', '매매평단가']]
display(dfk)

apt_col = dfk.아파트.unique()
ret = []

for aptName in apt_col:
    ret.append((dfk['아파트'] == aptName).sum())
```

```python
# 하대원동의 아파트별 매매평단가와 매매건수를 출력하세요.
df_m['매매건수'] = dfk

dfk = df_m[df_m['법정동'] == '하대원동']
dfk['아파트','매매평단가','매매건수'].groupby(by=['아파트'], observed=True).agg('mean')
```
