### 【 기술통계/요약통계 - 데이터의 특성을 수치로 설명하는 것 】
[1] 모듈 로딩 및 데이터 준비 <hr>

```python
## 모듈 로딩
import pandas as pd
import numpy as np
import wquantiles as wq    
```

```python
## 임의의 데이터
dataA = [70, 72, 75, 75, 76, 80, 82, 85, 88, 90]
dataSR = pd.Series(dataA)
```

[2] 데이터 정보확인 <hr>

```python
## 기술통계값 확인
dataSR.describe(percentiles=[0.3,0.6,0.9])
```

```python
## =========================================================================
## 평균/분산/표준편차 : pandas랑 numpy의 분산/표준편차 계산식의 차이로 값 다름
## =========================================================================
## -> pandas
print('Pandas 평균/분산/표준편차')
mean_pd = dataSR.mean()
print(f'평균    : {mean_pd}')
var_pd = dataSR.var(ddof=0)
print(f'분산    : {var_pd}')
std_pd = dataSR.std(ddof=0)
print(f'표준편차 : {std_pd}')
print()

## -> numpy
print('Numpy 평균/분산/표준편차')
mean_np = np.mean(dataSR)
print(f'평균    : {mean_np}')
var_np = np.var(dataSR)
print(f'분산    : {var_np}')
std_np = np.std(dataSR)
print(f'표준편차 : {std_np}')
```
